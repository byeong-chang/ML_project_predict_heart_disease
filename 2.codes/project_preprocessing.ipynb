```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
# preprocessing
df = pd.read_csv('./../0.data/heart_2020_cleaned.csv')

# replace 를 위한 함수 dataframe, column = 바꾸고자 하는 열 , text = 바꿀 문자열을 리스트로 넣어주면 list개수만큼 숫자로 변환해준다.
def exchange(DataFrame , column , *text):
    for i, t in enumerate(text[0]):
        DataFrame[column].replace(t,i,inplace = True)
# Yes or No 로 대답한 데이터들과 Female or Male 로 대답한 값에 대해 No = 0 Yes = 1로 replace 한 구분
# Diabetic은 예외적으로 4가지 경우에 대해서 올라와있었기에 0,1,2,3으로 카테고리화함   
exchange(df,'HeartDisease',['No','Yes'])
exchange(df,'Smoking',['No','Yes'])
exchange(df,'AlcoholDrinking',['No','Yes'])
exchange(df,'Stroke',['No','Yes'])
exchange(df,'DiffWalking',['No','Yes'])
exchange(df,'Sex',['Female','Male'])
exchange(df,'PhysicalActivity',['No','Yes'])
exchange(df,'Asthma',['No','Yes'])
exchange(df,'KidneyDisease',['No','Yes'])
exchange(df,'SkinCancer',['No','Yes'])
exchange(df,'Diabetic',['No','Yes','No, borderline diabetes','Yes (during pregnancy)'])
```

```python
from sklearn.preprocessing import LabelEncoder
item = [ '18-24', '25-29', '30-34', '35-39','40-44','45-49', '50-54', '55-59', '60-64','65-69', '70-74','75-79','80 or older']
encoder = LabelEncoder()
encoder.fit(item)
labels = encoder.transform(df['AgeCategory'])
df['AgeCategory'] = pd.Series(labels)
df
```

```python
item = ['White', 'Black', 'Asian', 'American Indian/Alaskan Native','Other', 'Hispanic']
encoder = LabelEncoder()
encoder.fit(item)
labels = encoder.transform(df['Race'])
df['Race'] = pd.Series(labels)
df['Race'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
item = df['GenHealth'].unique()
encoder = LabelEncoder()
encoder.fit(item)
labels = encoder.transform(df['GenHealth'])
df['GenHealth'] = pd.Series(labels)
df
```
